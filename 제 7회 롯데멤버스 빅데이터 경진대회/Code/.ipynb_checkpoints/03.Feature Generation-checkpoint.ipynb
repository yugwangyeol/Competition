# Feature Generation

```python
pdde_list = ['A01','A02','A03','A04','A05','A06']
cop_u_list = ['B01','C01','C02','D01','D02','E01']
cop_c_list = ['A01','A02','A03','A04','A05','A06','B01','C01','C02','D01','D02','E01']
```

```python
## PD_CLAC
def pd_unique(data):
    # 상품 구매 다양성 소분류, 중분류, 대분류
    f = data.query('cop_c in @pdde_list').groupby('cust')['clac_mcls_nm'].agg([('mpd_nunique','nunique')]).reset_index()
    return f

def favorite_pd(data):
    # 중분류 주구매 상품
    f = data.query('cop_c in @pdde_list').reset_index(drop=True).groupby('cust')['clac_mcls_nm'].agg(
        [('favorite_pd',lambda x: x.value_counts().index[0])]).reset_index().fillna('정보없음')
    return f
```

```python
## BR
def favorite_zon_loc(data):
    # 주구매지점 / 주구매지역
    f1 = data.groupby('cust')['zon_mcls'].agg([('favorite_zon', lambda x: x.value_counts().index[0])]).reset_index()
    f2 = data.groupby('cust')['zon_hlv_store'].agg([('favorite_loc',lambda x: x.value_counts().index[0])]).reset_index()
    f = pd.merge(f1,f2,how='outer')
    return f
```

```python
## 통합 관점
def purchase(data):
    # 총구매액, 구매건수, 평균구매액, 최대구매액
    f = data.groupby('cust')['buy_am'].agg(
        [('mean_buy_am',lambda x: np.round(np.mean(x))),('max_buy_am','max')]).reset_index()
    return f

def num_visit(data):
    # 내점일수
    f = data.groupby('cust')['de_dt'].agg([('num_visit','nunique')]).reset_index()
    return f

def day_sum_buy_am(data):
    # 내점 당 구매액 = 총구매액 / 내점일수
    visits = data.groupby('cust')['de_dt'].agg(lambda x: x.nunique()).reset_index()
    f = data.groupby('cust')['buy_am'].agg([('day_sum_buy_am','sum')]).reset_index()
    f['day_sum_buy_am'] = f['day_sum_buy_am']/visits['de_dt']
    return f

def day_count_buy_am(data):
    # 내점 당 구매건수
    visits = data.groupby('cust')['de_dt'].agg(lambda x: x.nunique()).reset_index()
    f = data.groupby('cust')['buy_am'].agg([('day_count_buy_am','count')]).reset_index()
    f['day_count_buy_am'] = f['day_count_buy_am']/visits['de_dt']
    return f

def days_(data):
    # 월초, 월중반, 월말별 구매 평균 금액, 구매 건수
    def days(x):
        if pd.to_datetime(x,format='%Y%m%d').day <= 10:
            return '월초'
        elif pd.to_datetime(x,format='%Y%m%d').day <= 20:
            return '월중반'
        else:
            return '월말'

    data['days'] = data['de_dt'].apply(days)
    f1 = pd.pivot_table(columns='days',index='cust',values='buy_am',data=data,aggfunc='mean',fill_value=0).reset_index()
    f1.columns = ['cust','buy_mean_end','buy_mean_middle','buy_mean_begin']
    f2 = pd.pivot_table(columns='days',index='cust',values='buy_am',data=data,aggfunc='size',fill_value=0).reset_index()
    f2.columns = ['cust','buy_count_end','buy_count_middle','buy_count_begin']
    f = pd.merge(f1,f2,how='outer')
    return f

def buy_cycle(data):
    # 구매 주기
    visits = data.groupby('cust')['de_dt'].agg(lambda x: x.nunique()).reset_index()
    de_max = data.groupby('cust')['de_dt'].agg([('buy_last','max')]).reset_index()
    de_min = data.groupby('cust')['de_dt'].agg([('buy_first','min')]).reset_index()
    de = de_max.merge(de_min)
    de['buy_cycle'] = de['buy_last'] - de['buy_first']
    de['num_visit'] = visits['de_dt']
    de['buy_cycle'] = de['buy_cycle']/de['num_visit']
    f = de[['cust','buy_cycle']]
    return f
    
def visit_weekend_ratio(data):
    # 주말 구매 비율
    f = data.groupby('cust')['de_dt'].agg(
        [('visit_weekend_ratio',lambda x: np.mean(pd.to_datetime(x,format='%Y%m%d').dt.dayofweek>4))]).reset_index()
    return f
    
def weekends(data):
    # 요일별 평균 구매 금액, 구매 건수
    def fw(x):
        return(('mon', 'tue', 'wed', 'thu', 'fri', 'sat', 'sun')[pd.to_datetime(x,format="%Y%m%d").weekday()])

    data['weekday'] = data['de_dt'].apply(fw)
    
    f1 = pd.pivot_table(columns='weekday',index='cust',values='buy_am',data=data,aggfunc='mean',fill_value=0).reset_index()
    f1.columns = ['cust','fri_mean','mon_mean','sat_mean','sun_mean','thu_mean','tue_mean','wed_mean']
    
    f2 = pd.pivot_table(columns='weekday',index='cust',values='buy_am',data=data,aggfunc='size',fill_value=0).reset_index()
    f2.columns = ['cust','fri_count','mon_count','sat_count','sun_count','thu_count','tue_count','wed_count']
    
    f = pd.merge(f1,f2,how='outer')
    return f
    
def most_buying(data):
    # 가장 구매를 많이 한 일자
    f1 = data.groupby('cust')['de_dt'].agg([('most_buying_day',lambda x: x.value_counts().index[0])]).reset_index()
    
    # 가장 구매를 많이 한 시간
    f2 = data.groupby('cust')['de_hr'].agg([('most_buying_hour',lambda x: x.value_counts().index[0])]).reset_index()
    
    # 가장 구매를 많이 한 요일
    def fw(x):
        return(('mon', 'tue', 'wed', 'thu', 'fri', 'sat', 'sun')[pd.to_datetime(x,format="%Y%m%d").weekday()])
    
    data['weekday'] = data['de_dt'].apply(fw)
    f3 = data.groupby('cust')['weekday'].agg([('most_buying_weekday',lambda x: x.value_counts().index[0])]).reset_index()
    
    f = pd.merge(f1,f2,how='outer').merge(f3,how='outer')
    return f
```

```python
def most_expensive(data):
    f1 = data.groupby('cust')['buy_am'].agg('max').reset_index()
    f = f1.merge(master12_tr[['cust','buy_am','clac_hlv_nm']],on=['cust','buy_am'],how='left')
    f.drop_duplicates('cust',inplace=True)
    f.drop(columns=['buy_am'],inplace=True)
    f.columns = ['cust','most_expensive']
    return f
```

```python
# 엘페이 오프라인 핫타임 구매량 / 구매수
def offline_hottime_buy_am(master_df):
    
    tp_df = master_df[(master_df['chnl_dv'] == 1)]
    f1 = pd.DataFrame(
        tp_df[(tp_df['de_hr'] == 17)|(tp_df['de_hr'] == 18)|(tp_df['de_hr'] == 19)].groupby('cust').buy_am.sum()).rename(
        columns= {'buy_am': 'off_hot_time_buy_am'}).reset_index()
    f2 = pd.DataFrame(
        tp_df[(tp_df['de_hr'] == 17)|(tp_df['de_hr'] == 18)|(tp_df['de_hr'] == 19)].groupby('cust').buy_ct.count()).rename(
        columns= {'buy_ct': 'off_hot_time_buy_count'}).reset_index()
    
    f = pd.merge(f1,f2,how='left').fillna(0)
    return f
```

```python
# 엘페이 온라인 핫타임 구매량 / 구매수
def online_hottime_buy_am(master_df):
    tp_df = master_df[(master_df['chnl_dv'] == 2)]
    f1 = pd.DataFrame(
        tp_df[(tp_df['de_hr'] == 10)|(tp_df['de_hr'] == 11)|(tp_df['de_hr'] == 12)].groupby('cust').buy_am.sum()).rename(
        columns= {'buy_am': 'on_hot_time_buy_am'}).reset_index()
    f2 = pd.DataFrame(
        tp_df[(tp_df['de_hr'] == 10)|(tp_df['de_hr'] == 11)|(tp_df['de_hr'] == 12)].groupby('cust').buy_ct.count()).rename(
        columns= {'buy_ct': 'on_hot_time_buy_count'}).reset_index()

    f = pd.merge(f1,f2,how='left')
    return f
```

```python
# 분기별 고객 구매시 평균 기온
def temperature(master_df):
    f = master_df.groupby('cust')['temperature'].agg([('mean_temperature',lambda x : np.mean(x.astype('float')))]).reset_index()
    return f
```

```python
# 분기별 고객 구매시 평균 습도
def huminity(master_df):
    f = master_df.groupby('cust')['huminity'].agg([('mean_huminity','mean')]).reset_index()
    return f
```

```python
# 분기별 고객 구매시 평균 강수량
def rainfall(master_df):
    f = master_df.groupby('cust')['rainfall'].agg([('mean_rainfall','mean')]).reset_index()
    return f
```

```python
def atmosphere(master_df):
    f = master_df.groupby('cust')['atmosphere'].agg([('mean_atmosphere','mean')]).reset_index()
    return f
```

```python
# 채널 다양성
def chnl_diversity(df):
    df = df.query('cop_c in @pdde_list').reset_index(drop=True).groupby('cust')['chnl_dv'].nunique().reset_index()
    df.columns = ['cust','pdde_chnl_diversity']
    df['pdde_chnl_diversity'] = df['pdde_chnl_diversity'].astype(int)
    return df
```

```python
# 채널 사용 횟수
def pdde_number_of_chnl(df):
    cc_df = df.query('cop_c in @pdde_list').reset_index(drop=True).groupby(['cust','chnl_dv'])['rct_no'].count().reset_index()
    cc_pt = cc_df.pivot('cust','chnl_dv','rct_no').fillna(0)#.reset_index(drop=True)
    cc_pt.columns = ['pdde_number_of_chnl_off','pdde_number_of_chnl_on']
    return cc_pt
```

```python
# 최다 사용 채널
def most_use_chnl(df):
    muc_df = df.groupby('cust')['chnl_dv'].agg([('pdde_most_use_chnl',lambda x: x.value_counts().index[0])]).reset_index()
    muc_df.loc[muc_df['pdde_most_use_chnl'] == 1,'pdde_most_use_chnl'] = "오프라인"
    muc_df.loc[muc_df['pdde_most_use_chnl'] == 2,'pdde_most_use_chnl'] = "온라인"
    return muc_df
```

```python
# 유통사 이용 다양성
def pdde_cop_c_utilization_diversity(df):
    pcu_df = df.query('cop_c in @pdde_list').reset_index(drop=True).groupby('cust')['cop_c'].nunique().reset_index()
    pcu_df.columns = ['cust','pdde_cop_c_utilization_diversity']
    return pcu_df
```

```python
# 주 사용 유통사
def pdde_cop_c_most_use(df):
    pc_df = df.query('cop_c in @pdde_list').groupby('cust')['cop_c'].agg(
        [('pdde_cop_c_most_use',lambda x: x.value_counts().index[0])]).reset_index()
    return pc_df
```

```python
# 구매 상품 다양성
def pdde_diversity_pd_c(df):
    return df.query('cop_c in @pdde_list').reset_index(drop=True).groupby('cust')['pd_c'].agg([('pd_c_difference','nunique')]).reset_index()
```

```python
# cop_u 사용 채널 다양성
def cop_u_chnl_diversity(df):
    df = df.query('cop_c not in @pdde_list').reset_index(drop=True).groupby('cust')['chnl_dv'].nunique().reset_index()
    df.columns = ['cust','cop_u_chnl_diversity']
    df['cop_u_chnl_diversity'] = df['cop_u_chnl_diversity'].astype(int)
    return df
```

```python
# 채널 사용 횟수
def cop_u_number_of_chnl(df):
    cc_df = df.query('cop_c not in @pdde_list').reset_index(drop=True).groupby(['cust','chnl_dv'])['rct_no'].count().reset_index()
    cc_pt = cc_df.pivot('cust','chnl_dv','rct_no').fillna(0)#.reset_index(drop=True)
    cc_pt.columns = ['cop_u_number_of_chnl_off','cop_u_number_of_chnl_on']
    return cc_pt
```

```python
# 제휴사 이용 다양성
def cop_u_cop_c_utilization_diversity(df):
    pcu_df = df.query('cop_c not in @pdde_list').reset_index(drop=True).groupby('cust')['cop_c'].nunique().reset_index()
    pcu_df.columns = ['cust','cop_u_cop_c_utilization_diversity']
    return pcu_df
```

```python
# 주 사용 제휴사
def cop_u_cop_c_most_use(df):
    pc_df = df.query('cop_c in @cop_u_list').groupby('cust')['cop_c'].agg(
        [('cop_u_cop_c_most_use',lambda x: x.value_counts().index[0])]).reset_index()
    return pc_df
```

```python
# 유통사 & 제휴사 이용 횟수
def pc_use(master_df):
    mq_gr = master_df.groupby(['cust','cop_c'])['rct_no'].count().reset_index()
    mq_pv = mq_gr.pivot('cust','cop_c','rct_no').reset_index().fillna(0)
    return mq_pv
```

```python
# 유통사 제휴사 총 이용수
def sum_use_pdde(data):
    mq_gr = data.groupby(['cust','cop_c'])['rct_no'].count().reset_index()
    mq_pv = mq_gr.pivot('cust','cop_c','rct_no').reset_index().fillna(0)
    f = pd.DataFrame(mq_pv.cust)
    f = pd.concat([f,mq_pv[pdde_list].sum(axis=1)],axis=1)
    f.columns = ['cust','sum_use_pdde']
    return f

def sum_use_cop_u(data):
    mq_gr = data.groupby(['cust','cop_c'])['rct_no'].count().reset_index()
    mq_pv = mq_gr.pivot('cust','cop_c','rct_no').reset_index().fillna(0)
    f = pd.DataFrame(mq_pv.cust)
    f = pd.concat([f,mq_pv[['cust']+cop_u_list].sum(axis=1)],axis=1)
    f.columns = ['cust','sum_use_cop_u']
    return f
```

```python
# 롯데 총 이용수
def sum_use(df):
    f = df.groupby('cust')['rct_no'].agg([('sum_use','count')]).reset_index()
    return f
```

```python
# RFM Score
def RFM_SCORE(df, start_date_choice=0, weights = [15, 35, 50]):
    start_date_list = [20210101,20210401,20210701,20211001]
    start_date = start_date_list[start_date_choice]
    
    recency = (df.groupby('cust')['de_dt'].max() - start_date).reset_index()
    recency.columns = ['cust','recency']
    
    frequency = df.groupby('cust')['de_dt'].count().reset_index()
    frequency.columns = ['cust','frequency']
    
    monetary = df.groupby('cust')['buy_am'].sum().reset_index()
    monetary.columns = ['cust','monetary']
    
    rfm_df = pd.merge(recency,frequency, on='cust').merge(monetary, on='cust')
    
    rfm_df['R'] = pd.qcut(rfm_df['recency'], q=5, labels=range(1,6)).astype(int)
    rfm_df['F'] = pd.qcut(rfm_df['frequency'], q=5, labels=range(1,6), duplicates='drop').astype(int)
    rfm_df['M'] = pd.qcut(rfm_df['monetary'], q=5, labels=range(1,6)).astype(int)
    
    rfm_df['RFMscore'] = weights[0] * rfm_df['R'] + weights[1] * rfm_df['F'] + weights[2] * rfm_df['M']
    
    return rfm_df
```

```python
# 쉬는 날
def weekend_holiday_use(df):
    f = df.groupby('cust')['weekend_holiday'].agg([('weekend_holiday_use','count')]).reset_index()
    return f
```

```python
def buy_ct_sum(data):
    f = data.query('buy_ct != "정보없음"').groupby('cust')['buy_ct'].agg([('buy_ct_sum',lambda x: sum(x.astype('float')))]).reset_index()
    return f
```

```python
def most_buy_cop_u(data):
    f = data.query('cop_c in @cop_u_list').groupby('cust')['buy_am'].agg('max').reset_index()
    f = f.merge(master12_tr[['cust','buy_am','cop_c']],on=['cust','buy_am'],how='left')
    f.drop(columns=['buy_am'],inplace=True)
    f.drop_duplicates('cust',inplace=True)
    f.columns = ['cust','most_buy_cop_u']
    return f
```

# Feature Make

```python
def feature_engineering(data):
    feature = [pd_unique(data),favorite_pd(data),favorite_zon_loc(data),purchase(data),num_visit(data),day_sum_buy_am(data),
               day_count_buy_am(data),days_(data),buy_cycle(data),visit_weekend_ratio(data),weekends(data),most_buying(data),
               offline_hottime_buy_am(data),online_hottime_buy_am(data),temperature(data),huminity(data),rainfall(data),atmosphere(data),
               chnl_diversity(data),pdde_number_of_chnl(data),most_use_chnl(data),pdde_cop_c_utilization_diversity(data),
               pdde_cop_c_most_use(data),pdde_diversity_pd_c(data),cop_u_chnl_diversity(data),cop_u_number_of_chnl(data),
               cop_u_cop_c_utilization_diversity(data),cop_u_cop_c_most_use(data),pc_use(data),sum_use_pdde(data),
               sum_use_cop_u(data),sum_use(data),RFM_SCORE(data),weekend_holiday_use(data),most_expensive(data),buy_ct_sum(data),
               most_buy_cop_u(data)]
    ## 피쳐 완성~
    features = pd.DataFrame({'cust': data.cust.unique()})
    for f in feature :
        features = pd.merge(features, f, on='cust', how='left')
    features[['favorite_zon','favorite_loc','most_buying_weekday','pdde_most_use_chnl','pdde_cop_c_most_use','cop_u_cop_c_most_use',
              'favorite_pd','most_expensive','most_buy_cop_u']].fillna('정보없음',inplace=True)
    features = features.fillna(0)
    features = features.merge(data[['cust','lpay_use_status']].drop_duplicates(),how='left')
    display(features)
    return features
```

```python
master12_tr_f = feature_engineering(master12_tr)
master23_tr_f = feature_engineering(master23_tr)
master3_te_f = feature_engineering(master3_te)
master4_te_f = feature_engineering(master4_te)
```
