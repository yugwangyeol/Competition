# Import

```python
import pandas as pd
from pycaret.classification import *
from sklearn.metrics import roc_auc_score
from sklearn import preprocessing
```

# DataLoad

```python
features12_tr = pd.read_csv('../3.로티로리_데이터 및 모델 세이브 파일/features12_tr.csv')
features3_te = pd.read_csv('../3.로티로리_데이터 및 모델 세이브 파일/features3_te.csv')
```

```python
print(features12_tr.shape)
display(features12_tr)
```

```python
print(features3_te.shape)
display(features3_te)
```

```python
features12_tr.drop('cust',axis=1,inplace=True)
features3_te.drop('cust',axis=1,inplace=True)
```

# Pycaret

```python
exp_clf = setup(data = features12_tr, target = 'lpay_use_status', session_id=123)
```

```python
best_model = compare_models(sort = 'AUC')
```

# Ensemble Model

```python
best_model_5 = compare_models(sort = 'AUC',n_select=5)
blender_5 = blend_models(best_model_5)
```

```python
final_model = finalize_model(blender_5)
```

```python
prediction = predict_model(final_model, data = features3_te)
```

```python
prediction
```

```python
ldf = pd.get_dummies(prediction['Label'])
ldf['Lpay_Offline'] = ldf['Lpay_Offline']*prediction['Score']
ldf['Lpay_On/Offline'] = ldf['Lpay_On/Offline']*prediction['Score']
ldf['Lpay_Online'] = ldf['Lpay_Online']*prediction['Score']
ldf['Lpay_X'] = ldf['Lpay_X']*prediction['Score']
ldf
```

```python
ldf['ms'] = 1-prediction['Score']
ldf['msv'] = ldf['ms']/4
```

```python
ldf
```

```python
ldf.loc[ldf['Lpay_Offline']==0,'Lpay_Offline'] = ldf['msv']
ldf.loc[ldf['Lpay_On/Offline']==0,'Lpay_On/Offline'] = ldf['msv']
ldf.loc[ldf['Lpay_Online']==0,'Lpay_Online'] = ldf['msv']
ldf.loc[ldf['Lpay_X']==0,'Lpay_X'] = ldf['msv']
```

```python
ldf = ldf.iloc[:,:4]
ldf
```

```python
ldf.iloc[:,0] = ldf.iloc[:,0]+(1-ldf.sum(axis=1))
```

```python
test_target = features3_te['lpay_use_status']
roc_auc_score(test_target,ldf,multi_class='ovr',average='weighted')
```
