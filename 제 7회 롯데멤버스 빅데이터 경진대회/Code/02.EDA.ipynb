# Data Load

```python
experience_df = pd.read_csv("../3.로티로리_데이터 및 모델 세이브 파일/Master DF.csv")
lpay_df = pd.read_csv("../3.로티로리_데이터 및 모델 세이브 파일/Refine lpay DF.csv")
```

```python
def chnl_change(x):
    if x == 1:
        return '오프라인'
    else:
        return '온라인'
```

```python
experience_df['chnl_dv'] = experience_df['chnl_dv'].apply(lambda x: chnl_change(x))
lpay_df['chnl_dv'] = lpay_df['chnl_dv'].apply(lambda x: chnl_change(x))
```

### Percentage Of Passengers

```python
def divide_chnl(x):
    if len(x) == 1:
        if x[0] == 1:
            return '오프라인'
        else:
            return '온라인'
    else:
        return '온/오프라인'
```

```python
tp_df_2 = pd.merge(experience_df, experience_df.groupby('cust')['chnl_dv'].unique().reset_index().rename(columns = {'chnl_dv': 'chnl_dv_byCust'}),on = 'cust', how='left')
tp_df_2['chnl_dv_byCust'] = tp_df_2['chnl_dv_byCust'].apply(divide_chnl)
```

```python
lpay_df = pd.merge(lpay_df, lpay_df.groupby('cust')['chnl_dv'].unique().reset_index().rename(columns = {'chnl_dv': 'chnl_dv_byCust'}),on = 'cust', how='left')
lpay_df['chnl_dv_byCust'] = lpay_df['chnl_dv_byCust'].apply(divide_chnl)
```

```python
tp_df_3 = lpay_df[(lpay_df['cop_c'] == 'B01')|(lpay_df['cop_c'] == 'E01')]
```

```python
ratio_tab = pd.crosstab(tp_df_3['cop_c'], tp_df_3['chnl_dv_byCust'], margins = True, normalize = 'index')
ratio_tab
```

```python
ratio_tab.iloc[:2, :].plot.bar(stacked = True, legend = False, color = ['dodgerblue' , '#fd8d49' ,'#ffe36e'])

plt.title('B01 / E01 제휴사 온라인/온오프라인/오프라인 이용객 비율')
plt.legend(bbox_to_anchor=(1.0, 1.0))
plt.ylabel('비율')
plt.show()
```
