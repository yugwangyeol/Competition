```python
# rle 디코딩
import os
import pandas as pd
import numpy as np
import cv2

def rle_to_mask(rle, image_size):
    mask = np.zeros(image_size[0] * image_size[1], dtype=np.uint8)
    rle = rle.split()

    for i in range(0, len(rle), 2):
        start = int(rle[i]) - 1
        length = int(rle[i + 1])
        end = start + length
        mask[start:end] = 255

    mask = mask.reshape((image_size[1], image_size[0])).T
    return mask

# csv 파일 읽기
csv_path = "/content/drive/MyDrive/patch16_train/patch16_train.csv"
dataframe = pd.read_csv(csv_path)

# 열(column) 값 추출
img_ids = dataframe["img_id"].tolist()
mask_rles = dataframe["mask_rle"].tolist()

# 폴더 경로 설정
save_folder = "/content/drive/MyDrive/patch16_train_mask"  # 이미지 저장할 폴더 경로 설정
os.makedirs(save_folder, exist_ok=True)  # 폴더가 없을 경우 생성

# 마스크 이미지 생성 및 저장
for img_id, mask_rle in zip(img_ids, mask_rles):
    try:
        image_size = (256, 256)
        if mask_rle != "-1":
            # 이미지 크기 설정 (예시로 256x256 설정)
            mask = rle_to_mask(mask_rle, image_size)
        else:
            mask = np.zeros(image_size, dtype=np.uint8)
            #print(mask_rle)
        filename = f"{img_id}_instance_color_RGB.png"  # 저장할 파일명 설정
        save_path = os.path.join(save_folder, filename)  # 저장할 폴더 경로와 파일명 결합
        cv2.imwrite(save_path, mask)
        print(f"Saved mask image: {save_path}")
    except:
        print(img_id, mask_rle)
```

```python
import os
import pandas as pd
import numpy as np
import cv2

def rle_to_mask(rle, image_size):
    mask = np.zeros(image_size[0] * image_size[1], dtype=np.uint8)
    rle = rle.split()

    for i in range(0, len(rle), 2):
        start = int(rle[i]) - 1
        length = int(rle[i + 1])
        end = start + length
        mask[start:end] = 255

    mask = mask.reshape((image_size[1], image_size[0])).T
    return mask

# csv 파일 읽기
csv_path = "/content/drive/MyDrive/국민의힘DATA/patch16_train.csv"
dataframe = pd.read_csv(csv_path)

#for k in ["train","val","test"]:
for k in ["train"]:
    # 이미지 파일 경로 설정
    image_folder = "/content/drive/MyDrive/국민의힘DATA2/"+k+"/images"  # 이미지 폴더 경로 설정

    # 마스크 이미지를 저장할 폴더 경로 설정
    save_folder = "/content/drive/MyDrive/국민의힘DATA2/"+k+"/labels"  # 마스크 이미지 저장할 폴더 경로 설정
    os.makedirs(save_folder, exist_ok=True)  # 폴더가 없을 경우 생성

    # 이미지 파일 이름들 가져오기
    image_files = os.listdir(image_folder)

    # 마스크 이미지 생성 및 저장
    for img_id, mask_rle in zip(dataframe["img_id"], dataframe["mask_rle"]):
        if f"{img_id}.png" in image_files:  # 이미지 파일이 존재하는 경우에만 처리
            try:
                image_size = (256, 256)
                if mask_rle != "-1":
                    mask = rle_to_mask(mask_rle, image_size)
                else:
                    mask = np.zeros(image_size, dtype=np.uint8)

                filename = f"{img_id}.png"  # 저장할 파일명 설정
                save_path = os.path.join(save_folder, filename)  # 저장할 폴더 경로와 파일명 결합
                cv2.imwrite(save_path, mask)
                print(f"Saved mask image: {save_path}")
            except:
                print(f"Error processing image: {img_id}")
        else:
            print(f"Image file not found: {img_id}.png")
```

```python
import os
import cv2

def check_and_delete_invalid_png_files(folder, target_folder):
    # 폴더 내의 모든 파일들을 순회하면서 PNG 파일인지 확인하고 오류가 있는지 검사
    for filename in os.listdir(folder):
        file_path = os.path.join(folder, filename)

        if filename.lower().endswith(".png"):
            try:
                # 이미지를 읽어오기 전에 예외 처리를 해서 오류를 방지
                img = cv2.imread(file_path)
                if img is None:
                    # 이미지가 제대로 읽히지 않으면 해당 파일을 삭제
                    os.remove(file_path)
                    print(f"Deleted {file_path}")
                    # 같은 이름을 가진 target_folder의 파일도 삭제
                    target_file_path = os.path.join(target_folder, filename)
                    if os.path.exists(target_file_path):
                        os.remove(target_file_path)
                        print(f"Deleted {target_file_path}")
            except Exception as e:
                # 오류가 발생하면 해당 파일을 삭제
                os.remove(file_path)
                print(f"Deleted {file_path}")
                # 같은 이름을 가진 target_folder의 파일도 삭제
                target_file_path = os.path.join(target_folder, filename)
                if os.path.exists(target_file_path):
                    os.remove(target_file_path)
                    print(f"Deleted {target_file_path}")

# 특정 폴더에서 오류가 발생한 PNG 파일들을 검사하고 삭제
source_folder = "/content/drive/MyDrive/국민의힘DATA2/train/images"  # 특정 폴더 경로 설정

# 특정 다른 폴더에서도 오류가 발생한 PNG 파일들을 검사하고 삭제
target_folder = "/content/drive/MyDrive/국민의힘DATA2/train/labels"  # 특정 다른 폴더 경로 설정
check_and_delete_invalid_png_files(source_folder, target_folder)
#Deleted /content/drive/MyDrive/국민의힘DATA/train/images/TRAIN_4953_8.png
#--------------------------------------------------------------------------
```

```python
%cd /content/drive/MyDrive/building-footprint-segmentation
import os
import sys
import json
import albumentations as A

module_path = os.path.abspath(os.path.join('../'))
if module_path not in sys.path:
    sys.path.append(module_path)

from building_footprint_segmentation.segmentation import init_segmentation, read_trainer_config
from building_footprint_segmentation.helpers.callbacks import CallbackList, load_callback
from building_footprint_segmentation.trainer import Trainer
```

```python
segmentation = init_segmentation("binary")
```

```python
augmenters = A.Compose([
    A.HorizontalFlip(p=0.5),
    A.RandomBrightnessContrast(p=0.2)
])
```

```python
model = segmentation.load_model(name="ReFineNet")
criterion = segmentation.load_criterion(name="BinaryCrossEntropy")
loader = segmentation.load_loader(
    root_folder="/content/drive/MyDrive/국민의힘DATA2",
    image_normalization="divide_by_255",
    label_normalization="binary_label",
    augmenters=augmenters,
    batch_size=32,
)
metrics = segmentation.load_metrics(
    data_metrics=["accuracy", "precision", "f1", "recall", "iou"]
)

optimizer = segmentation.load_optimizer(model, name="Adam")

callbacks = CallbackList()
# Ouptut from all the callbacks caller will be stored at the path specified in log_dir
for caller in  ["TrainChkCallback", "TimeCallback", "TensorBoardCallback", "TrainStateCallback"]:
    callbacks.append(load_callback(r"/content/drive/MyDrive/building-footprint-segmentation/model", caller))
```

```python
trainer = Trainer(
    model=model,
    criterion=criterion,
    loader=loader,
    metrics=metrics,
    callbacks=callbacks,
    optimizer=optimizer,
    scheduler=None,
)
```

```python
trainer.train(start_epoch=0, end_epoch=50)
```

```python
path_to_saved_state = r"/content/drive/MyDrive/국민의힘DATA2/model.pt"

# if new_end_epoch is None, the trainer will resume and run till end_epoch, the value of which is saved in state file
trainer.resume(state=path_to_saved_state, new_end_epoch=5)
```
